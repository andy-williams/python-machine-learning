# matplotlib

matplotlib is a data plotting library for python. It is heavily used in the field of data science.

Let's have a look at some of the most common ways matplotlib is used to visualise some data.

One thing to note is that the design of matplotlib seems to be procedural with a global state - this can catch you off guard if you're used to OO or functional way of doing things. This implies that every line of code that calls anything within matplotlib may invoke something that changes the global state of the next graph that we plot.

## Line Graph

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-6, 6, 0.001) # create a list starting at -6, incrementing to 6 with increments of 0.001

# https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot
# x, y
plt.plot(x, norm.pdf(x))
plt.show()
```

## Multiple

```python
plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))

plt.show()
```

## Grid

```python
axes = plt.axes()

axes.set_xlim([-6, 6])
axes.set_ylim([0, 1])

axes.set_xticks(np.arange(-6, 6, 1))
axes.set_yticks(np.arange(0, 1, 0.1))

axes.grid() # sets the grid for the next graph that is drawn

plt.plot(x, norm.pdf(x))
plt.plot(x, norm.pdf(x, 1.0, 0.5))

plt.show()
```

## Change Lines and Colours

```python
axes = plt.axes()

axes.set_xlim([-6, 6])
axes.set_ylim([0, 1])

axes.set_xticks(np.arange(-6, 6, 1))
axes.set_yticks(np.arange(0, 1, 0.1))

axes.grid() # sets the grid for the next graph that is drawn

# check https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot to see what b- and r: means
# they are used to set the colour and the style of line used
plt.plot(x, norm.pdf(x), 'b-')
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:')

plt.show()
```

## Labels and Legend

```python
axes = plt.axes()

axes.set_xlim([-6, 6])
axes.set_ylim([0, 1])

axes.set_xticks(np.arange(-6, 6, 1))
axes.set_yticks(np.arange(0, 1, 0.1))

axes.grid() # sets the grid for the next graph that is drawn


plt.xlabel('x')
plt.xlabel('y')

plt.plot(x, norm.pdf(x), 'b-')
plt.plot(x, norm.pdf(x, 1.0, 0.5), 'r:')

# https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.legend
plt.legend(['blue', 'red'], loc=2)

plt.show()
```

## Pie Chart

```python
values = [5, 5, 5, 5, 5, 5]
colors = ['r', 'g', 'b', 'c', 'm', 'b']
explode = [0, 0, 0.5, 0, 0, 0]
labels = ['Slice 1', 'Slice 2', 'Slice 3', 'Slice 4', 'Slice 5', 'Slice 6']

# http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.pie
plt.pie(values,
    colors = colors,
    labels = labels,
    explode = explode)
plt.title('Pizza')
plt.show()
```

## Bar Chart

```python
values = np.arange(10, 60, 10)

# https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.legend
plt.bar(range(1,6), values)
plt.show()
```

## Scatter Plot

```python
from numpy.random import randn

x = randn(200)
y = randn(200)

# https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.legend
plt.scatter(x, y)
plt.show()
```

## Histogram

```python
values = randn(500)

# https://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist
plt.hist(values, 100)
plt.show()
```
